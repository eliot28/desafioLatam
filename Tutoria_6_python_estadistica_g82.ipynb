```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv("tested.csv")
data
```

```python
# X = edad, clase y sexo
# y = murio o no
```

```python
data.columns
```

```python
data['Pclass'].value_counts()
```

```python
df_modelo = data[['PassengerId','Age','Pclass','Sex','Survived']]
df_modelo
```

```python
df_modelo.info()
```

```python
df_modelo = df_modelo.dropna()
```

```python
df_modelo.info()
```

```python
df_modelo.describe()
```

```python
df_modelo2 = pd.get_dummies(df_modelo, columns=['Sex'],drop_first=True)
df_modelo2['Sex_male'] = df_modelo2['Sex_male'].replace({True:1,False:0})
df_modelo2
```

```python
df_modelo2.columns
```

```python
X = df_modelo2[['Age','Pclass','Sex_male']].values
y = df_modelo2['Survived'].values
```

```python
df_modelo2['Survived'].value_counts()
```

```python
X
```

```python
y
```

```python
sacaler = StandardScaler()
X_s = sacaler.fit_transform(X)
X_s
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_s, y)
```

```python
len(X_resampled) ,len(X)
```

```python
pd.DataFrame(y_resampled).value_counts()
```

```python
len(y_resampled), len(y)
```

```python
regresion_logistic = LogisticRegression(random_state=42)
regresion_logistic.fit(X_resampled,y_resampled)
```

```python
## predicción
y_pred = regresion_logistic.predict(X)
```

```python
### calcular metricas
accuracy = accuracy_score(y,y_pred)
precision = precision_score(y,y_pred)
recall = recall_score(y,y_pred)
f1 = f1_score(y,y_pred)
roc_auc = roc_auc_score(y,y_pred)
PR = average_precision_score(y,y_pred)
```

```python
len(y)
```

```python
len(y_pred)
```

```python
print("Métricas de Desempeño:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("AUC-ROC:", roc_auc)
print("AUC-PR:", PR)
```
