```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
np.random.seed(42)
X =np.random.rand(100, 1) #Variable independiente
y =2*X + 1 + np.random.randn(100, 1) # Variable dependiente con ruido
data = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})
data.head(5)
```

```python
type(X)
```

```python
type(y)
```

```python
type(X.flatten())
```

```python
X.flatten()
```

```python
# Añadir una constante a la matriz de características
X = sm.add_constant(X)
X
```

```python
type(X)
```

```python
# Crear el modelo OLS (Ordinary Least Squares)
model = sm.OLS(y, X).fit()
# Imprimir un resumen del modelo
print(model.summary())
```

```python
# Obtener las predicciones del modelo
predictions = model.predict(X)
predictions
```

```python
# Calcular el Error Cuadrático Medio (MSE)
mse = ((y - predictions) ** 2).mean()
print(f'MSE: {mse}')
```

```python
# Calcular el Error Cuadrático Medio Explicado (EMSE)
emse = mse / ((y - y.mean()) ** 2).mean()
print(f'EMSE: {emse}')
```

```python
r_cuadrado = model.rsquared
print(f'r cuadrado: {r_cuadrado}')
```

```python
r_cuadrado_ajustado = model.rsquared_adj
r_cuadrado_ajustado
```

```python
# Calcular el Error Absoluto Medio (MAE)
mae = abs(y - predictions).mean()

print(mae)
```

```python
# Crear nuevos datos para predecir
new_X = np.array([[0],[1], [4]])
print(new_X)
new_X = sm.add_constant(new_X)
print(new_X)

```

```python
# Realizar predicciones
predictions = model.predict(new_X)
print(predictions)
```

```python
# Graficar los residuos
import matplotlib.pyplot as plt
residuals = model.resid
predictions = model.predict(X)
plt.scatter(data['X'], residuals)
plt.xlabel('X')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos')
#plt.plot(X, predictions, color='red', label="Línea de regresión")
plt.show()
```
