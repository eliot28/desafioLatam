a. Calcula el promedio, mediana y moda de las variables Altura (Height) y peso (Weight)

```
pokemon.columns
```

```
print('Promedio '+str(pokemon['Height'].mean()))
print('mediana '+str(pokemon['Height'].median()))
print('moda '+str(pokemon['Height'].mode()))
```

```
moda = pokemon['Height'].mode()
moda
```

```
print('Promedio '+str(pokemon['Weight'].mean()))
print('mediana '+str(pokemon['Weight'].median()))
print('moda '+str(pokemon['Weight'].mode()))
```

b. Calcula los quintiles, rango intercuartil y varianza de las mismas variables

```
## quintiles
primer_quintil = pokemon['Weight'].quantile(0.2)
segundo_quintil = pokemon['Weight'].quantile(0.4)
tercer_quintil = pokemon['Weight'].quantile(0.6)
cuarto_quintil = pokemon['Weight'].quantile(0.8)

print(primer_quintil)
print(segundo_quintil)
print(tercer_quintil)
print(cuarto_quintil)
```

```
## rango intercuantil

primer_rango = pokemon['Weight'].quantile(0.25)
segundo_rango = pokemon['Weight'].quantile(0.75)

print(primer_rango)
print(segundo_rango)
```
