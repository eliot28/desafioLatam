A)Calcula el promedio, mediana y moda de las variables Altura (Height) y peso
(Weight)


```python
pokemon[['Height','Weight']].describe()
```

```python
promedio_altura = pokemon['Height'].mean()
mediana_altura = pokemon['Height'].median()
moda_altura = pokemon['Height'].mode()

print(f'promedio altura es :{round(promedio_altura,2)}')
print(f'mediana altura es :{round(mediana_altura,2)}')
print(f'moda altura es :{round(moda_altura,2)}')
```

```python
promedio_peso = pokemon['Weight'].mean()
mediana_peso = pokemon['Weight'].median()
moda_peso = pokemon['Weight'].mode()

print(f'promedio peso es :{round(promedio_peso,2)}')
print(f'mediana peso es :{round(mediana_peso,2)}')
print(f'moda peso es :{round(moda_peso,2)}')
```

C)Calcula el promedio de peso y altura de los Pokemon, agrupados por Shapes

```python
tabla=pokemon.groupby('Shape')[['Height','Weight']].mean()
tabla = tabla.sort_values(by = 'Height', ascending= False)
tabla
```
