##<strong><font color="green">Preparación de datos</font></strong>
La preparación de datos es un paso crítico e indispensable en cualquier proceso de análisis de datos y visualización. En el mundo de los datos hay una regla muy conocida: el 80% del tiempo de un científico de datos se dedicó a preparar sus datos, y sólo el 20% a trabajar en ellos. La importancia de la preparación de los datos radica en varios aspectos:

*  **Calidad de los resultados:** los datos limpios, completos y bien estructurados conducen a resultados más precisos y confiables en el análisis. Si los datos están sucios o contienen errores, es probable que los resultados sean inexactos y conduzcan a conclusiones erróneas.

*  **Toma de decisiones informadas:** la preparación adecuada de datos permite a los analistas obtener una comprensión más profunda de ellos antes de realizar cualquier análisis o visualización. Esto ayuda a tomar decisiones informadas y fundamentadas en evidencia.

*  **Eficiencia en el análisis:** al preparar los datos de manera adecuada, se reduce el tiempo necesario para realizar el análisis. La limpieza y transformación previa facilitan el proceso y permiten que los analistas se centren en la interpretación y el descubrimiento de patrones relevantes.

*  **Asegurar consistencia:** la preparación de datos garantiza que los datos se presenten de manera coherente y homogénea. Esto es especialmente importante cuando se combinan datos de diferentes fuentes o cuando se trabaja en proyectos a largo plazo.

*  **Compatibilidad con herramientas y algoritmos:** muchas herramientas y algoritmos requieren que los datos estén en un formato específico. La preparación adecuada asegura que los datos se ajusten a los requisitos de las herramientas y algoritmos seleccionados.

*  **Identificación de errores y valores atípicos:** durante la preparación de datos, se pueden detectar y manejar valores atípicos, errores y datos inconsistentes. Esto es fundamental para evitar que estos problemas afecten negativamente el análisis y las conclusiones.

*  **Facilitar la visualización:** los datos bien preparados son más fáciles de visualizar y comunicar de manera efectiva. Una vez que los datos están limpios y estructurados, se pueden crear gráficos y representaciones visuales que sean claras y comprensibles.

*  **Minimizar sesgos y errores de interpretación:** la preparación de datos rigurosa y exhaustiva ayuda a minimizar el riesgo de sesgos y errores en la interpretación. Al abordar los problemas de calidad de datos, se asegura que el análisis sea más objetivo y confiable.


###<strong><font color="green">Etapas de la preparación de datos</font></strong>

La preparación de datos considera una serie de procesos y tareas que se llevan a cabo para garantizar que estén limpios, completos y listos para su análisis o visualización. Los principales procesos incluidos en la preparación de datos son los siguientes:

*  **Adquisición de datos:** este proceso involucra la recopilación de datos desde diversas fuentes, como bases de datos, archivos CSV, hojas de cálculo, APIs, fuentes web, sensores, entre otros.

*  **Exploración de datos (EDA - Exploratory Data Analysis):** es una etapa inicial en la que se examinan los datos para comprender su estructura, características y distribución. Se utilizan técnicas descriptivas y gráficos para obtener información preliminar.

*  **Limpieza de datos:** es el proceso de identificar y corregir errores, valores atípicos (outliers) y datos faltantes o nulos en el conjunto de datos. También puede implicar la estandarización de formatos y la corrección de inconsistencias.

*  **Transformación de datos:** en esta etapa se realizan cambios en los datos para que estén en un formato adecuado para el análisis o la visualización. Esto puede incluir la normalización de variables, la codificación de datos categóricos y la agregación de información.

* **Integración de datos:** cuando se trabaja con múltiples fuentes de datos, es necesario combinarlos de manera adecuada para crear un conjunto de datos coherente y completo.

*  **Selección de características (feature selection):** en el análisis de datos, en ocasiones es necesario reducir la cantidad de características o variables a las más relevantes para el objetivo específico del análisis.

*  **Reducción de dimensionalidad (dimensionality reduction):** si el conjunto de datos tiene muchas variables, esta etapa busca reducir la cantidad de dimensiones para facilitar el análisis y la visualización.

*  **División de datos:** en muchos casos, se divide el conjunto de datos en conjuntos de entrenamiento y prueba para el desarrollo y validación de modelos de machine learning.

*  **Creación de variables derivadas:** en algunos casos, es útil crear nuevas variables que sean combinaciones o transformaciones de las variables originales para obtener una perspectiva diferente.

*  **Formateo de datos para visualización:** preparar los datos en el formato adecuado para que las herramientas de visualización puedan interpretarlos y generar gráficos o representaciones visuales.

*  **Validación y verificación:** verificar que los datos preparados sean coherentes con las expectativas y que no haya errores en el proceso de preparación.

*  **Documentación:** registrar y documentar los pasos realizados en la preparación de datos para asegurar la reproducibilidad y facilitar el trabajo futuro.

Estos procesos no son necesariamente lineales, y en la práctica, pueden requerir iteraciones y ajustes a medida que se avanza en el análisis y se descubren nuevos desafíos en los datos.

###<strong><font color="green">Datos duplicados</font></strong>

Los datos duplicados son observaciones que aparecen más de una vez en el conjunto de datos. Estos duplicados pueden surgir debido a errores en la recopilación o el ingreso de datos, y pueden distorsionar el análisis y las conclusiones.

Para detectar duplicados, es necesario comparar las observaciones y buscar coincidencias exactas o casi exactas en todas las variables relevantes. Una vez identificados, los duplicados se pueden eliminar para garantizar que cada observación sea única y no afecte negativamente el análisis.

Python nos ofrece el método **drop.duplicates**, cuya sintaxis es:

<center>
data_sin_duplicados = df.drop_duplicates(subset=[‘columna1’,’columna2’, ... ])
</center>


De esta manera, guardaremos en un nuevo dataFrame el original (**df**) sin los datos repetidos  **considerando las columnas indicadas**. Es decir, el método considerará un dato como duplicado si coinciden los datos correspondientes a las columnas especificadas, aunque no estén repetidos en las otras. Si no pondemos nada, el método considerará como dato repetido solo si la coincidencia es absoluta, es decir, en todas las columnas.

Observa, además, que si aplicamos el método "directamente", es decir, sin guardar en otro dataFrame, el resultado se guardará en el dataFrame original, es decir, lo modificará.

###<strong><font color="green">Transformación de datos</font></strong>

Esta etapa implica aplicar técnicas para ajustar los datos a un formato adecuado que facilite su interpretación y análisis. Entre las transformaciones más comunes se encuentran la normalización y estandarización de datos, la discretización y binning de variables numéricas, y la codificación de variables categóricas.


#### Normalización y estandarización de datos:

La normalización y estandarización son técnicas que se utilizan para escalar los datos numéricos y colocarlos en una escala común.  Se trata de ajustar los valores para que estén en un rango específico, generalmente entre 0 y 1, lo que es útil cuando se trabaja con diferentes unidades o magnitudes. Por otro lado, la estandarización transforma los datos para que tengan una media de 0 y una desviación estándar de 1, lo que ayuda a eliminar sesgos y comparar variables con diferentes escalas.

La normalización y estandarización son especialmente importantes cuando se utilizan algoritmos de machine learning que se basan en distancias o magnitudes, como el gradiente descendente o el algoritmo K-Nearest Neighbors (K-NN).

#### Discretización y binning de variables numéricas

Es el proceso de convertir variables numéricas en variables categóricas al agrupar los valores en rangos o intervalos. Esta técnica es útil cuando se quiere simplificar el análisis o cuando se tienen datos continuos que se pueden agrupar en categorías significativas.

Por ejemplo, si se tiene una variable de edades, se puede discretizar en grupos como "jóvenes", "adultos" y "ancianos". El binning es una forma específica de discretización que divide el rango de valores en intervalos equitativos, lo que facilita la visualización y el análisis.


#### Codificación de variables categóricas

Las variables categóricas representan atributos con diferentes categorías o etiquetas, como género, tipo de producto o país. Los algoritmos de machine learning suelen requerir que los datos estén en formato numérico, por lo que es necesario codificar las variables categóricas en valores numéricos.

Existen varias técnicas de codificación, como la codificación one-hot, donde se crea una columna binaria para cada categoría, o la codificación ordinal, que asigna valores numéricos a las categorías según un orden específico.

La transformación de datos juega un papel fundamental en el análisis y modelado de datos, ya que ayuda a garantizar que los datos sean adecuados para su uso con diversas técnicas y algoritmos. Al aplicar técnicas de normalización, estandarización, discretización y codificación, los datos se vuelven más comprensibles y aptos para la generación de conocimientos significativos y la toma de decisiones informadas.


##<strong><font color="green">Análisis exploratorio de datos (EDA)</font></strong>

El Análisis Exploratorio de Datos (EDA) es una fase crucial en cualquier proceso de análisis y visualización de datos. En esta etapa, los analistas utilizan una combinación de estadísticas descriptivas y gráficos para obtener una comprensión profunda de los datos y extraer información relevante. Dos aspectos fundamentales del EDA son el uso de estadísticas descriptivas y la exploración de relaciones y correlaciones entre variables.

### Uso de estadísticas descriptivas y gráficos para entender la distribución de datos

Las estadísticas descriptivas son herramientas clave en el EDA que resumen y caracterizan los datos de manera concisa. Estas incluyen medidas como la media, mediana, moda, desviación estándar, mínimo, máximo y cuartiles. Estas estadísticas permiten entender la tendencia central, la dispersión y la forma de la distribución de los datos.

Además de las estadísticas descriptivas, los gráficos desempeñan un papel crucial en el EDA, ya que permiten visualizar la distribución de los datos de manera más intuitiva. Algunos gráficos comunes en el EDA incluyen histogramas para visualizar la distribución de variables numéricas, gráficos de barras para representar frecuencias de variables categóricas, y gráficos de caja (boxplots) para mostrar estadísticas resumidas y detectar valores atípicos.

### Exploración de relaciones y correlaciones entre variables

El análisis exploratorio de relaciones y correlaciones es una parte esencial del EDA. Consiste en examinar cómo se relacionan las variables entre sí y cómo influyen mutuamente en el conjunto de datos. Esto se puede lograr utilizando gráficos de dispersión (scatter plots) para visualizar la relación entre dos variables numéricas.

Además, la correlación es una métrica que cuantifica la relación lineal entre dos variables numéricas. Un coeficiente de correlación cercano a 1 indica una correlación positiva fuerte, mientras que un valor cercano a -1 indica una correlación negativa fuerte. Una correlación cercana a 0 indica una relación débil o inexistente. Estos análisis permiten identificar patrones, tendencias y posibles dependencias entre variables, lo que puede ser fundamental para la toma de decisiones y la construcción de modelos predictivos.

###<strong><font color="green">Gráfico de línea</font></strong>

Vamos a graficar los consumos promedio por día en el restaurante. Para ello debemos primero escoger nuestros datos.

```python
# Agrupar los datos por la columna 'day' y calcular el valor total promedio para cada día
promedio_por_dia = tips.groupby('day')['total_bill'].mean()

print(promedio_por_dia)
```

Con esto, podemos graficar nuestros datos.

```python
#creamos un diccionario con los datos, y lo pasamos a un DataFrame

plt.figure(figsize=(8,6))  #configuramos el tamaño de la figura (8 x 6 pulgadas)
sns.lineplot(data=promedio_por_dia, marker='.' ,markersize=50, color='b', linestyle='-',label='Consumo promedio')
plt.xlabel('Días de la Semana')
plt.ylabel('Consumo promedio')
plt.title('Variación de Temperatura durante la Semana')
plt.legend()
# Mostrar el gráfico
plt.grid(True)  # Agregar cuadrícula de fondo (opcional)
plt.show()
```

####Observaciones

Hay parámetros que ya conoces, y otros que podemos precisar:

* **marker:** Especifica el tipo de marcador para los puntos de datos en la gráfica de líneas. Algunas opciones comunes son:

  *  '.': Punto
  *  ',': Píxel
  *  'o': Círculo
  *  '^': Triángulo hacia arriba
  *  'v': Triángulo hacia abajo
  *  's': Cuadrado
  *  '+': Cruz
  *  '*': Estrella
  *  'x': Cruz diagonal
  *  'D': Diamante
  *  'p': Pentágono

* **color:** Especifica el color de la línea. Puedes usar nombres de colores comunes como 'b' para azul, 'r' para rojo, 'g' para verde, entre otros. También se pueden utilizar códigos hexadecimales para colores personalizados como '#FF5733'.

*  **markersize:** Especifica el tamaño de los marcadores utilizados en la gráfica de líneas. Puedes proporcionar un valor numérico para establecer el tamaño deseado.

###<strong><font color="green">Histogramas</font></strong>

Seaborn nos ofrece dos funciones: **histplot** para una sola variable y crear un histograma simple, y **distplot** para más complejidad.

```python
data=tips['total_bill']
sns.histplot(data, bins=10, color='skyblue') #observa variando el valor de bins, o simplemente no ponerlos
plt.title('Título')
plt.xlabel('Consumo total')
plt.ylabel('Frecuencia')
plt.show()
```

Ahora, utilizando **distplot** para crear dos gráficos en uno:

¿Por qué usar distplot? Veremos que nos permite utilizar algunos otros parámetros y funcionalidades. Por ejemplo, podemos construir dos histogramas en uno separando por una variable categórica (en nuestro caso, 'diagnosis')

```python
#creamos primero un objeto
ax = sns.displot(data=tips, x='total_bill',  hue='time', fill=True)
#estamos señalando que, de los datos de "data", grafique los correspondientes a mean_perimeter distinguiendo por el valor de diagnosis

ax.set(xlabel='Consumo',ylabel='Frecuencia', title='Consumo por tipo de comida')
plt.show()
```

##<strong><font color="green">Swarmplot</font></strong>

Swarmplot se utiliza para visualizar la distribución de puntos para cada categoría individual en un conjunto de datos categóricos.


```python
sns.set(style='whitegrid')
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('Swarm Plot - Total de la Cuenta por Día')
plt.xlabel('Día de la Semana')
plt.ylabel('Total de la Cuenta')
# Mostrar el gráfico
plt.show()
```

####Observaciones

En este ejemplo, utilizamos sns.swarmplot para crear un gráfico de dispersión (swarm plot) que muestra cómo se distribuye el total de la cuenta (total_bill) en función del día de la semana (day). Cada punto representa un valor individual del total de la cuenta y se agrupa verticalmente según el día correspondiente.

##<strong><font color="green">Boxplot</font></strong>

Los boxplots son especialmente útiles para:
*  Identificar la simetría o sesgo en la distribución de los datos.
*  Detectar valores atípicos o valores extremos.
*  Comparar la distribución de varios conjuntos de datos.
*  Visualizar la dispersión y concentración de los datos.


```python
# Crear el boxplot para comparar el total de la cuenta (total_bill) según el día de la semana (day)
sns.boxplot(x='day', y='total_bill', data=tips)

# Personalización opcional
plt.title('Box Plot - Total de la Cuenta por Día')
plt.xlabel('Día de la Semana')
plt.ylabel('Total de la Cuenta')

# Mostrar el boxplot
plt.show()
```

##<strong><font color="green">Subplots</font></strong>

Podemos, finalmente, crear una visualización con diferentes gráficos organizados en la forma que queramos, gracias a **subplots**

```python
# Crear una figura con 3x2 subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 15))

# Gráfico 1 - Scatter plot (total_bill vs. tip)
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[0, 0])
axes[0, 0].set_title('Scatter Plot - Total Bill vs. Tip')

# Gráfico 2 - Box plot (day vs. total_bill)
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[0, 1])
axes[0, 1].set_title('Box Plot - Total Bill por Día')

# Gráfico 3 - Bar plot (sex vs. total_bill)
sns.barplot(x='sex', y='total_bill', data=tips, ax=axes[1, 0])
axes[1, 0].set_title('Bar Plot - Total Bill por Género')

# Gráfico 4 - Violin plot (day vs. total_bill)
sns.swarmplot(x='day', y='total_bill', data=tips, ax=axes[1, 1])
axes[1, 1].set_title('Violin Plot - Total Bill por Día')

# Gráfico 5 - Count plot (day)
sns.countplot(x='day', data=tips, ax=axes[2, 0]) #analiza qué hace este gráfico
axes[2, 0].set_title('Count Plot - Día')

# Gráfico 6 - Histograma (total_bill)
sns.histplot(x='total_bill', data=tips, kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Histograma - Total Bill')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar los subplots
plt.show()
```
