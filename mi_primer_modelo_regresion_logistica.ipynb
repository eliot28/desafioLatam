```python
## dejar variables importantes
train = train[['PassengerId','Survived','Pclass','Age','Sex']].dropna()
sex = pd.get_dummies(train['Sex'],drop_first=True)
train = pd.concat([train,sex],axis=1)
train['male'] = train['male'].replace({True:1,False:0})
train.head()
```

```python
train = train.drop(['Sex'],axis=1)
train
```

```python
train.info()
```

```python
train['Age'] = train['Age'].astype('int64')
```

```python
X=train.drop('Survived',axis=1)
y=train['Survived']


X_test  = test.drop("PassengerId",axis=1).copy()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrmodel = LogisticRegression() # Creamos la instancia
lrmodel.fit(X_train,y_train)
```

```python
prediccion=lrmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report # Nos indicara la precision de nuestro modelo
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,prediccion))
```
